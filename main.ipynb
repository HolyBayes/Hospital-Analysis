Задание 1 График числа потребления за весь год(потребление) от 3-х признаков: возраст, пол и семейное положение

```python
%pylab inline
import pylab as plt
import numpy as np

ages, sex = database['Age'].values(), database['Sex'].values()
family_statuses, visits = database['Family status'].values(), database['Total visits'].values()
woman_amount = sum(1 for value in sex if value == 'w')
woman_visits = sum(int(visits[i]) for i in xrange(len(visits)) if sex[i] == 'w')
woman_visit_factor = float(woman_visits) / woman_amount

men_amount = sum(1 for value in sex if value == 'm')
men_visits = sum(int(visits[i]) for i in xrange(len(visits)) if sex[i] == 'm')
man_visit_factor = float(men_visits) / men_amount

pylab.subplot(aspect=True)
pylab.pie([woman_visit_factor, man_visit_factor], labels=['Woman visit', 'Men visit'], autopct='%i%%')
```

Видно, что женщины чуть больше обращаются в данную клинику

```python
visits_from_age = {}
for i in xrange(len(ages)):
    if 0 <= ages[i] <= 1000:
        age_interval = int(ages[i] / 10)
        if visits_from_age.has_key(age_interval):
            visits_from_age[age_interval] += visits[i]
        else:
            visits_from_age.update({age_interval: visits[i]})
pylab.plot(map(lambda x: 10*x, visits_from_age.keys()), visits_from_age.values())
pylab.ylabel('Visits')
pylab.xlabel('Age')
```

По данному графику видно, что наибольшее число посетителей - дети до 10 лет и взрослые с 20 до 30 лет. Подростки не уделяют время своему здоровью вообще.

Создаем базу {семейное положение: [суммарное количество визитов, число человек]}.

```python
visits_from_family_status = {}
for i in xrange(len(family_statuses)):
    family_status = family_statuses[i]
    if family_status:
        if visits_from_family_status.has_key(family_status):
            visits_from_family_status[family_status][0] += visits[i]
            visits_from_family_status[family_status][1] += 1
        else:
            visits_from_family_status.update({family_status: [visits[i], 1]})
print visits_from_family_status
```

Невалидные ключи (семейные положения) спокойно отбрасываем - они составляют ничтожную долю от общей статистики. Второй повод тому - их невалидность. Например немудрено, что семейное положение "никакое"(а такие есть) будет достаточно малоинформативным. Для основных же семейных положений посчитаем коэффициент посещаемости как среднее число посещений на человека для каждой группы. Критерий отброса - число человек с таким с.п. > 10 и притом с.п. считается указанным

```python
vfsf = {item[0]: item[1] for item in visits_from_family_status.items() if item[1][1] > 10 and str(item[0]) != 'nan'}  #Visit-from-Family-Status-Filtered
print vfsf
```

```python
vffs = {item[0]: float(item[1][0]) / item[1][1] for item in vfsf.items()}    #Visit Factor from Family Status
pos = np.arange(len(vffs))
plt.title('Visit factor dependency from family status')
plt.bar(xrange(len(vffs)), vffs.values())
plt.xticks(map(lambda x: x + 0.5, xrange(len(vffs))), vffs.keys())
plt.ylabel('Visit factor')
```

На лицо корреляция между посещаемостью и семейным положением. Думаю, что комментировать, кто ходит к врачу чаще, руководствуясь полученной гистограммой - излишне.

Задание 3. Портрет «нашего» потребителя – кто больше всего платит и на что?

Будем отбирать <=10% самых щедро-платящих. Определим для начала денежную границу этих 10 процентов

```python
wasted_database = {}    #dictionary with {wasted sum: indexes} structure
for i in xrange(len(database['Total sum'])):
    wasted_sum = database['Total sum'][i]
    if wasted_database.has_key(wasted_sum):
        wasted_database[wasted_sum].append(i)
    else:
        wasted_database.update({wasted_sum: [i]})
wasted_rank = wasted_database.keys()
wasted_rank.sort(key=lambda x: -x)   #sort by desceding
top_wasting_amount = int(len(database['Total sum']) * 0.10)

last_top_user, amount, temporal_amount = 0, 0, 0
while temporal_amount < top_wasting_amount:
    last_top_user += 1
    amount = temporal_amount
    temporal_amount += len(wasted_database[wasted_rank[last_top_user]])
    
top_users = []
for i in xrange(last_top_user):
    for user in wasted_database[wasted_rank[i]]:
        top_users.append(user)

sex = [database['Sex'][i] for i in top_users]
pylab.title = "Top user's sex diagram"
pylab.subplot(aspect=True)
pylab.pie([sex.count('m'), sex.count('w')], labels = ["Men", "Woman"], autopct='%i%%')
```

```python
import collections
family_statuses = collections.Counter()
for i in top_users:    
    family_statuses[database['Family status'][i]] += 1
    
family_statuses.pop(nan, None)
for key in family_statuses.keys():
    if family_statuses[key] < 5:
        family_statuses.pop(key, None)
print family_statuses
```

```python
plt.title("Family statuses of top 10%")
plt.bar(xrange(len(family_statuses)), family_statuses.values())
plt.xticks(map(lambda x: x + 0.5, xrange(len(family_statuses))), family_statuses.keys())
plt.ylabel('People amount')
```

```python
age = collections.Counter()
for i in top_users:
    age[database['Age'][i]] += 1
plt.title("Age of top 10 %")
plt.bar(xrange(len(age)), age.values())
plt.ylabel('Amount')
plt.xlabel('Age')
```

Видно, что основные клиенты - это преимущественно женатые женщины средних лет и дети

Задание 4.	Выделить тех, кто много ходит, но мало тратит, какие еще у них особенности? 

Наиболее адекватная оценка - по среднему значению потраченных денег за визит на каждого человека (возьмем первые 10% начиная с самого низкого значения этого показателя)

```python
spend_ranks = {} #  {spend_rank: [id]}
for i in xrange(len(database['Age'])):#  len(database['Age']) ~ nrows
    spend_rank = database['Total sum'][i] / database['Total visits'][i]
    if spend_ranks.has_key(spend_rank):
        spend_ranks[spend_rank].append(i)
    else:
        spend_ranks.update({spend_rank: [i]})
bottom_factor = 0.10
bottom_spend_ranks = sorted(spend_ranks.keys())[:int(bottom_factor * len(spend_ranks.keys()))]
# print bottom_spend_ranks
greedy_id = []#  id of top 10% greediest users
for rank in bottom_spend_ranks:
    greedy_id.extend(spend_ranks[rank])
greedy_men = sum(1 for i in greedy_id if database['Sex'][i] == 'm')
greedy_woman = sum(1 for i in greedy_id if database['Sex'][i] == 'w')
pylab.pie([greedy_men, greedy_woman], labels = ["Men", "Woman"], autopct='%i%%')
```

Вывод на основе предыдущих гистограмм - женщины чаще обращаются, но по мелким услугам. Мужчины наоборот

```python
family_statuses = collections.Counter()
for i in greedy_id:    
    family_statuses[database['Family status'][i]] += 1
    
family_statuses.pop(nan, None)
for key in family_statuses.keys():
    if family_statuses[key] < 5:
        family_statuses.pop(key, None)
print family_statuses
```

```python
plt.title("Family statuses of bottom 10%")
plt.bar(xrange(len(family_statuses)), family_statuses.values())
plt.xticks(map(lambda x: x + 0.5, xrange(len(family_statuses))), family_statuses.keys())
plt.ylabel('People amount')
```

Задание 5.	Есть ли зависимость среднего чека от частоты хождения? (Например, те, кто часто ходят склонны и платить больше за один раз или нет?)

Средний чек был посчитан ранее как spend_rank = {spend_rank: [id]}

```python
from dateutil.parser import parse
fvf = [[], []]#  Frequency from Visit Factor
for i in xrange(len(database['Age'])):#  database['Age']~nrows
    spend_rank = database['Total sum'][i] / database['Total visits'][i]
    frequency = database['Total visits'][i] / ((parse(database['Last visit'][i]) - parse(database['First visit'][i])).days / 365 + 1)
    fvf[0].append(frequency)
    fvf[1].append(spend_rank)
plt.scatter(fvf[1], fvf[0])
axes = plt.gca()
axes.set_xlim(0, 8000)
plt.ylabel('Frequency')
plt.xlabel('Average ticket')
plt.title("Visit frequency from average ticket cost")
```

По этому графику можно сделать сразу несколько выводов: 1. В основном частота обращения мала. 2. В основном доминирует низкий средний чек. 3. Клиенты с большой частотой обращения обычно пользуются услугами с низким значением среднего чека. 4. Клиенты с высоким значением среднего чека, как правило, обращаются редко. Задание 8.	Есть ли зависимость между объемом трат за год  и «давностью»пациента (новенький или давно наш клиент)?

```python
sd = {}#  Spend from Duration
for i in xrange(len(database['Age'])):
    years_in_clinique = abs((parse(database['Last visit'][i]) - parse(database['First visit'][i])).days / 365) + 1
    spend = database['Total sum'][i] / years_in_clinique
    if sd.has_key(years_in_clinique):
        sd[years_in_clinique].append(spend)
    else:
        sd.update({years_in_clinique: [spend]})
for key in sd.keys():
    sd[key] = float(sum(sd[key])) / len(sd[key])

plt.title("Av.spend from client's duration")
plt.bar(xrange(len(sd)), sd.values())
plt.xticks(map(lambda x: x + 0.5, xrange(len(sd))), sd.keys())
plt.xlabel("Client's duration in clinique (years)")
plt.ylabel('Average spend')
```

Рост проглядывается достаточно четко. Во всяком случае, "одногодки" тратят значительно меньше постоянных клиентов. Возможно, связано с первоначальным доверием к организации.

Задание 9.	Есть ли зависимость объема потраченных денег от возраста? И какая?

```python
spend_from_age = {}
bin_width = 10
for i in xrange(len(database['Age'])):
    if str(database['Age'][i]) != 'nan':
        age = bin_width * int(database['Age'][i] / bin_width)
        spend = database['Total sum'][i]
        if spend_from_age.has_key(age):
            spend_from_age[age].append(database['Total sum'][i])
        else:
            spend_from_age.update({age: [spend]})
for key in spend_from_age.keys():
    spend_from_age[key] = float(sum(spend_from_age[key])) / len(spend_from_age[key])

age_bins = sorted(spend_from_age.keys())
spend_bins = [spend_from_age[age] for age in age_bins]
plt.title("Spend from age dependency")
plt.bar(xrange(len(age_bins)), spend_bins)
plt.xticks(map(lambda x: x + 0.5, xrange(len(age_bins))), age_bins)
plt.xlabel('Age group')
plt.ylabel('Spend')
```

Из гистограммы четко видна зависимость потраченной суммы от возраста. Больше всего за здоровьем следят дети (вернее их родители) и люди средних лет. Дальше идет постоянный спад с ростом возраста. Подростки (10 < age < 20) не следят за здоровьем.
